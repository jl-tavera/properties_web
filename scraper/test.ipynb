```python
import requests
from bs4 import BeautifulSoup
import json
import re

def extract_next_data(url):
    headers = {
        "User-Agent": "Mozilla/5.0"
    }
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    # Find the <script id="__NEXT_DATA__">
    script_tag = soup.find("script", id="__NEXT_DATA__")
    if not script_tag:
        return None

    data = json.loads(script_tag.string)
    return data.get("props", {}).get("pageProps", {}).get("data", {})

# Example usage
url = "https://www.fincaraiz.com.co/apartamento-en-arriendo-en-cable-manizales/10833459"
data = extract_next_data(url)
print(json.dumps(data, indent=2))
```
